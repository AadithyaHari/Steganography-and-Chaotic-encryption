```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("Automobile_data.csv")
df
```

```python
df.info()
```

# CHECKING FOR THE NULL VALUES-->

```python
newdf=df.replace("?",np.nan)
newdf.isnull().sum()
```

# REPLACING THE NULL VALUES----->

```python
newdf['normalized-losses']=newdf['normalized-losses'].astype(float)
summean=newdf['normalized-losses'].mean()
newdf['normalized-losses'].fillna(summean,inplace=True)


newdf['price']=newdf['price'].astype(float)
summean=newdf['price'].mean()
newdf['price'].fillna(summean,inplace=True)

newdf['bore']=newdf['bore'].astype(float)
summean=newdf['bore'].mean()
newdf['bore'].fillna(summean,inplace=True)

newdf['stroke']=newdf['stroke'].astype(float)
summean=newdf['stroke'].mean()
newdf['stroke'].fillna(summean,inplace=True)

newdf['horsepower']=newdf['horsepower'].astype(float)
summean=newdf['horsepower'].mean()
newdf['horsepower'].fillna(summean,inplace=True)

newdf['peak-rpm ']=newdf['peak-rpm'].astype(float)
summean=newdf['peak-rpm '].mean()
newdf['peak-rpm '].fillna(summean,inplace=True)

newdf
```

# DEPENDENCE OF CURB WEIGHT AND HIGHWAY MILEAGE--->

```python
gh=df[['curb-weight','highway-mpg']]
print(gh.sort_values(by='highway-mpg',ascending=False))
sns.jointplot(x='curb-weight',y='highway-mpg',data=df)
plt.tight_layout()
#Here in the jointplot we basically see the relation between any two columns in the dataset and see a trend between the two.
#Default kind of the plot is scatter but if we want to cahnge the kind we should type there kind='hex'..
#And kind='reg' stands for regression is also another plot...
#Another kind is kde which represents a 2 dimensional plot..
```

# RANGES OF LENGTH,WIDTH AND HEIGHT USING DISTRIBUTION PLOT--->

```python
fig,axes=plt.subplots(1,3,figsize=(20,10))
sns.distplot(df['length'],kde=False,bins=30,ax=axes[0])
sns.distplot(df['width'],kde=False,bins=30,ax=axes[1])
sns.distplot(df['height'],kde=False,bins=30,ax=axes[2])
plt.tight_layout()
#kde represents kernel density estimate which is by default plotted in the dist plot, but if we dont want the kde plot then we have to write as kde=false....
```

```python
sns.jointplot(x='length',y='height',data=df,kind='hex')
```

# HEAT MAP--->

```python
new=df[['length','width','height','city-mpg','highway-mpg']]
tc=new.corr()
sns.heatmap(tc,annot=True,cmap='coolwarm',linecolor='black',linewidth=1)
#corr() is a function by which we can make a datframe into a matrix form(means it this form all the rows and columns have the same headers..
#This is used in plotting the heat map which tells us about the values gradient relative to each other..
#Annot keyword is used to show all the values in the plot relative to each other...
#cmap represents the color of grid..
```
